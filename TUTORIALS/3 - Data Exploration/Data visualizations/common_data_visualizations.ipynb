-------------------------
### Import the required libraries

```python
import os
import sys

platform_path = os.path.abspath(os.path.join(os.path.abspath(''), '../../../'))
sys.path.append(platform_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

-------------------------
### Read data into a dataframe

```python
df = pd.read_csv(os.path.join(platform_path, 'DATA/recent-grads.csv'))
```

-------------------------
### Create a histogram for a variable

```python
df.hist(column=['Median'], grid=False, figsize=(10,5))
```

-------------------------
### Create histograms to compare two variables

```python
cols = ['Median', 'Sample_size']

df.hist(column=cols, layout=(len(cols),1), grid=False, figsize=(10,5*len(cols)))
```

-------------------------
### Create a set of box and whisker plots to compare multiple variables

```python
cols = ['Sample_size', 'Major_category']
group_by_col = 'Major_category'

ax = df.loc[:, cols].boxplot(by=group_by_col, figsize=(10,5))
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

-------------------------
### Create a scatter plot to compare two variables

```python
x_col = 'Median'
y_col = 'Unemployment_rate'

df.plot(kind='scatter', x=x_col, y=y_col, figsize=(8,6))
```

-------------------------
### Create a scatter plot with color overlay to compare three variables

```python
x_col = 'Median'
y1_col = 'Unemployment_rate'
y2_col = 'ShareWomen'

fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(x=df[x_col], y=df[y1_col], label=y1_col, color='r')    
ax.scatter(x=df[x_col], y=df[y2_col], label=y2_col, color='g')

ax.set_xlabel(x_col)
ax.set_ylabel(y1_col + ' / ' + y2_col)

ax.legend()
plt.show()
```
